```python
df = pd.read_csv("shopping_trends_updated.csv")
```

Датасет содержит набор данных о покупательских предпочтениях клиентов. Предлагает ценную информацию о поведении потребителей и моделях покупок. 

| Признак         | Описание                                              | Единицы измерения |
|-----------------|-------------------------------------------------------|-------------------|
| Gender          | Gender of the customer (Male/Female)                  |                   |
| Category        | Category of the item purchased                        |                   |
| Location        | Location where the purchase was made                  |                   |
| Review Rating   | Rating given by the customer for the purchased item   |                   |
| Shipping Type   | Type of shipping chosen by the customer               |                   |
| Payment Method  | Customer's most preferred payment method              |                   |

```python
df["Purchase Amount (USD)"] = df["Purchase Amount (USD)"].replace(np.linspace(1, 50, dtype=int), np.nan)
df.head(10)
```

```python
df["Purchase Amount (USD)"] = df["Purchase Amount (USD)"].replace(np.nan, df["Purchase Amount (USD)"].mean())
df.head(20)
```

```python
df_ATD = df.copy()

df_ATD["Purchase Amount (USD) to Review Rating"] = df["Purchase Amount (USD)"] / df["Review Rating"]
df_ATD.head(5)
```

```python
df_ATD["Purchase Amount (USD) to Review Rating"] = df.apply(lambda row: row["Purchase Amount (USD)"] / row["Review Rating"], axis=1)
df_ATD.head(5)
```

```python
df_ATD["Review Rating x 2"] = df["Review Rating"].apply(lambda x: x * 2)
df_ATD.head(5)
```

11. .merge

```python
df1 = pd.DataFrame({
    'student_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
})

df2 = pd.DataFrame({
    'student_id': [1, 2, 3, 5],
    'grade': ['A', 'B', 'C', 'A']
})

merged_df = pd.merge(df1, df2, on='student_id')

merged_df
```

12. .concat

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}, index=[0, 1, 2, 3])

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
}, index=[4, 5, 6, 7])

result = pd.concat([df1, df2])

result
```
